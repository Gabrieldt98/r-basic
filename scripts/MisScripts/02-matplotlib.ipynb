78. Múltiples gráficos en una misma figura

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5.0,0.1)
x2 = np.arange(0,5.0,0.2)
```

```python
plt.figure(1)
plt.subplot(211) # números se pueden separar con coma
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k')

plt.subplot(212)
plt.plot(x2, f(x2), 'g--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(2,1,1)
plt.plot([1,2,3])
plt.subplot(2,1,2)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(2,1,1)
plt.title("Esto es el primer título")
```

# Texto en los gráficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, facecolor = "g", alpha = 0.6)
plt.xlabel("Coeficiente Intelectual", fontsize = 14, color = "blue")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu, \sigma)$")
plt.text(120, 500, r'$\mu = 100,\ \sigma = 20$')
plt.axis([20,180,0,700])
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10,6), dpi = 90)
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x, y, lw = 2.0)

plt.annotate('Máximo local', xy = (4*np.pi,1), xytext = (15, 1.5),
            arrowprops = dict(facecolor = "black", shrink = 0.05))
plt.ylim(-2,2)
plt.show()
```

81. Cambios de ejes y escalas

```python
from matplotlib.ticker import NullFormatter
```

```python
import matplotlib.pyplot as plt
import numpy as np
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(figsize = (10,8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala lineal")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala logarítmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x, y - y.mean())
plt.yscale("symlog", linthresh = 0.1) # lintreshy se cambió a linthresh de acuerdo con warning de documentación
plt.title("Escala log simétrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Escala logística")
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)

plt.show()
```

# Cambios en los ejes

```python
import numpy as np
x = np.linspace(-np.pi, np.pi, 256, endpoint = True)
S, C = np.sin(x), np.cos(x)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,8))

plt.plot(x,S, color = "blue", linewidth = 1.2, linestyle = "-", label = "seno")
plt.plot(x,C, color = "green", linewidth = 1.2, linestyle = "-", label = "coseno")
#plt.xlim(-4,4)
#plt.ylim(-1.2,1.2)
# Límites de forma dinámica:
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)
# Divisiones que aparecen directamente en números enteros: ticks
#plt.xticks(np.linspace(-4,4, 9, endpoint = True))
#plt.yticks(np.linspace(-1,1, 5, endpoint = True))
# Crear vector en donde se quieren exactamente los ticks:
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) # en xyticks además de suministrar vector con valores, después de coma se coloca segundo vector con la expresión con la que quieres pintar cada uno de esos valores
plt.yticks(np.linspace(-1,1, 2, endpoint = True),
          ['-1', '+1'])

# Para ajustar posición de ejes dentro del gráfico, obtener los ejes del dibujo actual para modificarlo

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.legend(loc = "upper left")


x0 = 2*np.pi/3
plt.plot([x0,x0], [0,np.sin(x0)], color = 'blue', linewidth = 2.5, linestyle = '--')
plt.scatter([x0, ], [np.sin(x0), ], color = 'blue')
plt.annotate(r'$\sin(\frac{2 \pi}{3}) = \frac{\sqrt{3}}{2}$',
            xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (+20,+40), textcoords = "offset points",
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

plt.plot([x0,x0], [0,np.cos(x0)], color = 'green', linewidth = 2.5, linestyle = '--')
plt.scatter([x0, ], [np.cos(x0), ], color = 'green')
plt.annotate(r'$\cos(\frac{2 \pi}{3}) = -\frac{1}{2}$',
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90,-60), textcoords = "offset points",
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle = "arc3,rad=.2"))

for label in ax.get_xticklabels() + ax.get_yticklabels(): # obtengo el conjunto de todas las etiquetas del eje de las x u todas las etiquetas del eje de las y
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor = "white", edgecolor = "None", alpha = 0.6))
    
    
plt.show()
```
